```python
from neuralprophet import NeuralProphet
import pandas as pd
from matplotlib import pyplot as plt

# Load the data
df = pd.read_csv('Data/tidsserie_SIR1.csv')

# Create the model
m = NeuralProphet()

# Fit the model
m.fit(df, freq="D")

df_future = m.make_future_dataframe(df, periods=30)

# print last date of df with description "last day of df"
print(f"last day of df_future {df_future.tail(1)}")

# Make a forecast 
forecast = m.predict(df_future)

# print last date of forecast with description "last date of forecast"
print(f"last date of forecast {forecast.tail(1)}")

# Plot the forecast
fig_forecast = m.plot(forecast)
fig_components = m.plot_components(forecast)
fig_model = m.plot_parameters()
plt.show()





```
